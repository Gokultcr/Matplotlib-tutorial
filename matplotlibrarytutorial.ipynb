# Different types of Plottings in Matplot library

#### Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Reading the data

```python
data = pd.read_csv('../input/iris-flower-dataset/IRIS.csv')
```

### Data Analysis

## Graph Plotting

##### Line Plot

```python
sepal_length = data.sepal_length
sepal_width = data.sepal_width
petal_length = data.petal_length
species= data.species

plt.plot(sepal_length , sepal_width)
plt.show()
```

#### Ploting without line

```python
plt.plot(sepal_length , sepal_width, 'o')
plt.show()
```

##### Dotted line


* linestyle can be written as ls
* dotted can be written as :
* dashed can be written as --

```python
plt.plot(sepal_length , linestyle = 'dotted')
plt.show()
```

##### Line color

```python
plt.plot(sepal_length , linestyle = 'dotted', color = 'r')
plt.show()
```

##### Line width

```python
plt.plot(sepal_length ,color = 'g', linewidth = '10.5')
plt.show()
```

##### Colour list:

* 'b' -> Blue 
* 'c' -> Cyan
* 'm' -> Magenta
* 'y' -> Yellow
* 'k' -> Black
* 'w' -> White

#### Setting Font and position for Title

```python
plt.plot(sepal_length, color = 'r')
plt.plot(sepal_width, color = 'g')
font1 = {'family':'serif','color':'green','size':20}
font2 = {'family':'serif','color':'hotpink','size':15}
plt.title(" Iris Data set", loc = 'left', fontdict = font1)
plt.xlabel("Sepal Length", fontdict = font2)
plt.ylabel("Sepal Width", fontdict = font2)
plt.show()
```
